# Importation des bibliothèques

```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

Fonction permettant d'afficher une série temporelle



```
def affiche_serie(temps, serie, format="-", debut=0, fin=None, label=None):
    plt.plot(temps[debut:fin], serie[debut:fin], format, label=label)
    plt.xlabel("Temps")
    plt.ylabel("Valeur")
    if label:
        plt.legend(fontsize=14)
    plt.grid(True)
```

# Tendance et périodicité

Définition d'une fonction pour créer un courbe de tendance avec une pente que l'on peut choisir

```
def tendance(temps, pente=0):
    return pente * temps
```

Création d'une série temporelle avec la fonction précédente :

```
temps = np.arange(4 * 365 + 1)            # temps = [0 1 2 .... 4*365] = [0 1 2 .... 1460]
serie = tendance(temps, 0.1)              # serie = droite de pente 0.1

plt.figure(figsize=(10, 6))
affiche_serie(temps, serie)
plt.show()
```

Ajoutons maintenant un motif périodique à notre série. Tout d'abord, définissons le motif que nous allons répéter :

$motif = \left\{ {\begin{array}{*{20}{c}}
{\cos \left( {2\pi t} \right){\rm{       }}\qquad0 \le t < 0.4}\\
{{e^{ - 3t}}{\rm{                }}\qquad\qquad0.4 \le t \le 1}
\end{array}} \right.$

```
t = np.arange(1000)
t = t/1000
motif = np.where(t < 0.4,np.cos(t * 2 * np.pi),1 / np.exp(3 * t))

plt.plot(t[0:1000], motif[0:1000], "-", label="Motif")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.grid(True)
```

Définissons maintenant les fonctions pour ajouter le motif à notre courbe de tendance :  
- `motif_periodique`: Cette fonction créée le motif sur des instants t = [0,1]
- `ajoute_motif` : Cette fonction ajoute le motif à la série 

```
def motif_periodique(instants):
    return (np.where(instants < 0.4,                            # Si les instants sont < 0.4
                    np.cos(instants * 2 * np.pi),               # Alors on retourne la fonction cos(2*pi*t)
                    1 / np.exp(3 * instants)))                   # Sinon, on retourne la fonction exp(-3t)

def ajoute_motif(temps, periode, amplitude=1, phase=0):
    """Répétition du motif sur la même période"""
    instants = ((temps + phase) % periode) / periode            # Mapping du temps =[0 1 2 ... 1460] => instants = [0.0 ... 1.0]
    return amplitude * motif_periodique(instants)
```

```
amplitude = 40
serie = ajoute_motif(temps, periode=365, amplitude=amplitude)

plt.figure(figsize=(10, 6))
affiche_serie(temps, serie)
plt.show()
```

Construit la série temporelle en ajoutant une tendance à la répétition du motif précédent, avec une amplitude de 40 :



```
pente = 0.05
moyenne = 10
serie = moyenne + tendance(temps, pente) + ajoute_motif(temps, periode=365, amplitude=amplitude)

plt.figure(figsize=(10, 6))
affiche_serie(temps, serie)
plt.show()
```

# Bruit

Dans la réalité, les séries ne sont pas aussi jolies : Elles sont bruitées, avec parfois un rapport signal sur bruit très faible... Ajoutons donc maintenant bruit blanc à notre série.  
Pour cela, on utilise la fonction `randn` qui produit des valeurs à virgules flottantes, avec une fonction de distibution de type normale (gaussienne), ne valeur moyenne 0 et de variance 1.

```
def bruit_blanc(temps, amplitude_bruit=1, graine=None):
    rnd = np.random.RandomState(graine)
    return rnd.randn(len(temps)) * amplitude_bruit
```

Regardons à quoi ressemble ce bruit :

```
plt.figure(figsize=(10, 6))
bruit = bruit_blanc(temps,amplitude_bruit=1,graine=42)
affiche_serie(temps, bruit)
plt.show()
```

Affichons l'histogramme et une fonction de distribution de type gaussienne N(0,1) :

```
import scipy.stats as stats
import math as mth

variance = 1

hist, t = np.histogram(bruit)
plt.hist(bruit)
plt.plot(np.linspace(t[0],t[len(t)-1],100),stats.norm.pdf(np.linspace(t[0],t[len(t)-1],100),0,1)*max(hist)/((1/(variance*mth.sqrt(2*mth.pi)))))
```

On appelle ce genre de bruit un **buit blanc** car sa densité spectrale de puissance est identique à toutes les fréquences. Ceci correspond à une autocorrélation nulle en tout point sauf à l'origine : le processus est décorrélé. Nous allons voir ce qu'est l'autocorrélation dans un instant...

Mais avant, ajoutons ce bruit blanc à notre série :

```
serie = serie + bruit

plt.figure(figsize=(10, 6))
affiche_serie(temps, serie)
plt.show()
```

```
def impulses(time, num_impulses, amplitude=1, seed=None):
    rnd = np.random.RandomState(seed)
    impulse_indices = rnd.randint(len(time), size=10)
    series = np.zeros(len(time))
    for index in impulse_indices:
        series[index] += rnd.rand() * amplitude
    return series    
```

```
def autocorrelation(source, φs):
    ar = source.copy()
    max_lag = len(φs)
    for step, value in enumerate(source):
        for lag, φ in φs.items():
            if step - lag > 0:
              ar[step] += φ * ar[step - lag]
    return ar
```
